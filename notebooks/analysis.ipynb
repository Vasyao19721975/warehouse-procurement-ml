```python
# 1) импорт библиотек
import pandas as pd
```

```python
# 2) загрузка датасета
df = pd.read_csv("../data/processed/final_dataset.csv")
df.head()
```

```python
# 3) размер таблицы
df.shape
```

```python
# 4) названия столбцов
df.columns.tolist()
```

```python
# 5) проверка дат
df["date"] = pd.to_datetime(df["date"])
print("Минимальная дата:", df["date"].min())
print("Максимальная дата:", df["date"].max())
print("Количество уникальных дат:", df["date"].nunique())
```

```python
# 6) количество товаров
print("Количество уникальных SKU:", df["sku_id"].nunique())
print("Количество уникальных товаров:", df["product"].nunique())
```

```python
# 7) базовая статистика
df[["stock", "supply", "sales"]].describe()
```

```python
# 8) сколько продаж равно нулю
(df["sales"] == 0).sum()
```

```python
# 9) топ товаров по суммарным продажам
top_sales = (
    df.groupby(["sku_id", "product"], as_index=False)["sales"]
    .sum()
    .sort_values("sales", ascending=False)
)

top_sales.head(20)
```

```python
# 10) товары без продаж
zero_sales = (
    df.groupby(["sku_id", "product"], as_index=False)["sales"]
    .sum()
)

zero_sales = zero_sales[zero_sales["sales"] == 0]
zero_sales.head(20)
```

```python
# 11) продажа и сортировка
# считаем суммарные продажи по каждому товару
product_sales = (
    df.groupby(["sku_id", "product"], as_index=False)["sales"]
    .sum()
)

# сортируем
product_sales = product_sales.sort_values("sales", ascending=False)

product_sales.head(20)
```

```python
# 12) разделяем товары
def categorize_sales(x):
    if x == 0:
        return "no_sales"
    elif x < 5:
        return "slow"
    elif x < 20:
        return "medium"
    else:
        return "fast"

product_sales["category"] = product_sales["sales"].apply(categorize_sales)

product_sales["category"].value_counts()
```

```python
# 13) ищем опасные товары
# берём последнюю дату
latest_date = df["date"].max()

# остатки на последнюю дату
latest_stock = df[df["date"] == latest_date][["sku_id", "product", "stock"]].copy()

# объединяем с продажами
risk_products = pd.merge(
    latest_stock,
    product_sales[["sku_id", "product", "sales", "category"]],
    on=["sku_id", "product"],
    how="left"
)

# фильтр
risk_products = risk_products[
    (risk_products["category"].isin(["no_sales", "slow"])) &
    (risk_products["stock"] >= 10)
]

# сортировка
risk_products = risk_products.sort_values("stock", ascending=False)

risk_products.head(30)
```

```python
# 14) средние продажи за день
# считаем количество дней в данных
num_days = df["date"].nunique()

# добавляем средние продажи в день
product_sales["sales_per_day"] = product_sales["sales"] / num_days

product_sales.head()
```

```python
# 15) Опасные товары
risk_products = pd.merge(
    risk_products,
    product_sales[["sku_id", "sales_per_day"]],
    on="sku_id",
    how="left"
)

risk_products.head()
```

```python
# 16) дни покрытия
risk_products["days_of_stock"] = risk_products["stock"] / (risk_products["sales_per_day"] + 0.0001)

risk_products.head()
```

```python
# 17) сортировка
risk_products = risk_products.sort_values("days_of_stock", ascending=False)

risk_products.head(30)
```

```python
# 18) формирование рекомендаций по закупке
def purchase_decision(days):
    if days > 30:
        return "no_order"
    elif days > 10:
        return "reduce_order"
    elif days > 5:
        return "normal_order"
    else:
        return "increase_order"

risk_products["decision"] = risk_products["days_of_stock"].apply(purchase_decision)

risk_products.head(30)
```

```python
# 19) расчет рекомендуемого количества закупки

# желаемый запас (например 7 дней)
target_days = 7

# расчет
risk_products["recommended_order"] = (
    risk_products["sales_per_day"] * target_days
) - risk_products["stock"]

# если отрицательное → 0
risk_products["recommended_order"] = risk_products["recommended_order"].apply(lambda x: max(0, x))

risk_products.head(30)
```

```python
# 20) рекомендации для всех товаров

# берём последние остатки
latest_date = df["date"].max()

all_products = df[df["date"] == latest_date][["sku_id", "product", "stock"]].copy()

# добавляем продажи
all_products = pd.merge(
    all_products,
    product_sales[["sku_id", "sales", "sales_per_day"]],
    on="sku_id",
    how="left"
)

# считаем дни покрытия
all_products["days_of_stock"] = all_products["stock"] / (all_products["sales_per_day"] + 0.0001)

# применяем решения
all_products["decision"] = all_products["days_of_stock"].apply(purchase_decision)

# считаем рекомендованную закупку
target_days = 7

all_products["recommended_order"] = (
    all_products["sales_per_day"] * target_days
) - all_products["stock"]

all_products["recommended_order"] = all_products["recommended_order"].apply(lambda x: max(0, x))

# сортируем по спросу
all_products = all_products.sort_values("sales_per_day", ascending=False)

all_products.head(30)
```

```python
# 21) распределение решений по закупке
all_products["decision"].value_counts()
```

```python
# 22) средняя рекомендуемая закупка по типу решения
all_products.groupby("decision")["recommended_order"].describe()
```

```python
# 23) подготовка данных для ML

ml_df = df.copy()

# сортируем по товару и дате
ml_df = ml_df.sort_values(["sku_id", "date"])

# добавляем прошлые продажи (lag)
ml_df["lag_1"] = ml_df.groupby("sku_id")["sales"].shift(1)
ml_df["lag_2"] = ml_df.groupby("sku_id")["sales"].shift(2)

# удаляем пустые строки
ml_df = ml_df.dropna()

ml_df.head()
```

```python
# 24) обучение ML модели

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

# признаки (вход)
X = ml_df[["lag_1", "lag_2"]]

# целевая переменная (что предсказываем)
y = ml_df["sales"]

# делим на обучение и тест
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# создаём модель
model = RandomForestRegressor(n_estimators=50, random_state=42)

# обучаем
model.fit(X_train, y_train)

print("Модель обучена")
```

```python
# 25) проверка модели

from sklearn.metrics import mean_absolute_error

# прогноз
y_pred = model.predict(X_test)

# ошибка
mae = mean_absolute_error(y_test, y_pred)

print("Средняя ошибка (MAE):", mae)
```

```python
# 26) прогноз продаж с помощью ML

ml_df["predicted_sales"] = model.predict(X)

ml_df.head()
```

```python
# 27) средний прогноз продаж по товару

predicted_sales = (
    ml_df.groupby("sku_id", as_index=False)["predicted_sales"]
    .mean()
)

predicted_sales.head()
```

```python
# 28) добавляем прогноз в all_products

all_products = pd.merge(
    all_products,
    predicted_sales,
    on="sku_id",
    how="left"
)

all_products.head()
```

```python
# 29) пересчёт закупки через ML

target_days = 7

all_products["ml_recommended_order"] = (
    all_products["predicted_sales"] * target_days
) - all_products["stock"]

all_products["ml_recommended_order"] = all_products["ml_recommended_order"].apply(lambda x: max(0, x))

all_products.head(30)
```

```python
# 30) сравнение обычной и ML-рекомендации

comparison = all_products[[
    "sku_id",
    "product",
    "stock",
    "sales_per_day",
    "predicted_sales",
    "recommended_order",
    "ml_recommended_order"
]].copy()

comparison["difference"] = comparison["recommended_order"] - comparison["ml_recommended_order"]

comparison.head(30)
```

```python
# 31) товары с наибольшим отличием между подходами

comparison.sort_values("difference", ascending=False).head(20)
```

```python
# 32) сохранение финального результата

all_products.to_csv("../outputs/final_recommendations.csv", index=False)
```
